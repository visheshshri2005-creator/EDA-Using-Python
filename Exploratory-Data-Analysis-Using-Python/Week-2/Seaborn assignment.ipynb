```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='darkgrid')
```

##  Load Datasets

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
flights = sns.load_dataset('flights')

tips.head()
```

##  Relational Plots

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', style='smoker')
plt.title('Total Bill vs Tip')
plt.show()
```

##  Categorical Plots

```python
sns.barplot(data=tips, x='day', y='tip')
plt.title('Average Tip by Day')
plt.show()
```

##  Distribution Plots

```python
sns.histplot(data=tips, x='total_bill', kde=True)
plt.title('Distribution of Total Bill')
plt.show()
```

##  Regression Plot

```python
sns.regplot(data=tips, x='total_bill', y='tip')
plt.title('Regression: Total Bill vs Tip')
plt.show()
```

```python
flights = sns.load_dataset('flights')

flights_pivot = flights.pivot(index="month", columns="year", values="passengers")

sns.set_theme(style="white")
sns.set_palette("coolwarm")

plt.figure(figsize=(12,6))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Monthly Passenger Numbers')
```

The heatmap will render properly, showing seasonal peaks in summer months and a clear upward trend in passenger numbers over the years.
