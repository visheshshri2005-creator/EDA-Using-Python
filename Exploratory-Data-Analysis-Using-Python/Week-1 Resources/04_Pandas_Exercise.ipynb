Welcome to a quick exercise for you to practice your pandas skills! We will be using the [SF Salaries Dataset](https://www.kaggle.com/kaggle/sf-salaries) from Kaggle! Just follow along and complete the tasks outlined in bold below. The tasks will get harder and harder as you go along.

**Import pandas as pd.**

```python
import pandas as pd
```

**Read Salaries.csv as a dataframe called sal.**

```python
sal = pd.read_csv('Salaries.csv')
```

**Check the head of the DataFrame.**

```python
sal.head()
```

**Use the .info() method to find out how many entries there are.**

```python
sal.info()
```

**What is the average BasePay ?**

```python
sal['BasePay'] = pd.to_numeric(sal['BasePay'], errors='coerce')

print(sal['BasePay'].mean())
```

**What is the highest amount of OvertimePay in the dataset ?**

```python
sal['OvertimePay'] = pd.to_numeric(sal['OvertimePay'], errors='coerce')
print(sal['OvertimePay'].max())
```

**What is the job title of  JOSEPH DRISCOLL ? Note: Use all caps, otherwise you may get an answer that doesn't match up (there is also a lowercase Joseph Driscoll).**

```python
a = sal.loc[sal['EmployeeName'] == "JOSEPH DRISCOLL", 'JobTitle'].tolist()
a
```

**How much does JOSEPH DRISCOLL make (including benefits)?**

```python
b = sal.loc[sal['EmployeeName'] == "JOSEPH DRISCOLL", 'TotalPayBenefits'].tolist()
b
```

**What is the name of highest paid person (including benefits)?**

```python
sal['TotalPay'] = pd.to_numeric(sal['TotalPay'], errors='coerce')
max_pay = sal['TotalPay'].max()
name = sal.loc[sal['TotalPay'] == max_pay, 'EmployeeName'].tolist()
name
```

**What is the name of lowest paid person (including benefits)? Do you notice something strange about how much he or she is paid?**

```python
sal['TotalPay'] = pd.to_numeric(sal['TotalPay'], errors='coerce')
min_pay = sal['TotalPay'].min()
name2 = sal.loc[sal['TotalPay'] == min_pay, 'EmployeeName'].tolist()
name2
```

**What was the average (mean) BasePay of all employees per year? (2011-2014) ?**

```python
sal.groupby('Year')['BasePay'].mean()
```

**How many unique job titles are there?**

```python
J = sal['JobTitle'].unique()
print(len(J))
```

**What are the top 5 most common jobs?**

```python
job_counts = sal.groupby('JobTitle').size().reset_index(name='count')
job_counts = job_counts[job_counts['JobTitle'].isin(J)]
job_counts = job_counts.sort_values('count', ascending=False)
print(job_counts.head().to_string(index=False))
```

**How many Job Titles were represented by only one person in 2013? (e.g. Job Titles with only one occurence in 2013?)**

```python
sal_2013 = sal[sal['Year'] == 2013]
job_counts_2013 = sal_2013.groupby('JobTitle').size().reset_index(name='count')
n = (job_counts_2013['count'] == 1).sum()
print(n)
```

**How many people have the word Chief in their job title? (This is pretty tricky)**

```python
mask = sal['JobTitle'].str.contains(r'\bChief\b', case=False, na=False)
chiefs = sal.loc[mask, 'EmployeeName'].unique().tolist()
print(len(chiefs))
```

**Is there a correlation between length of the Job Title string and Salary?**

```python
sal['title_len'] = sal['JobTitle'].str.len()
correlation = sal[['title_len', 'TotalPayBenefits']].corr()

print(correlation)
```
